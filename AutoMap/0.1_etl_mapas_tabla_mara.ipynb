# Preparacion de los datos para crear los mapas + DF VINCULADO 

```python
# Asegurarse de que todos los valores en 'Lugar_de_inicio' son cadenas
estacionados_cliente['Lugar_de_inicio'] = estacionados_cliente['Lugar_de_inicio'].astype(str)

# Separar la columna en dos nuevas columnas, manejando posibles errores
def split_coords(lugar):
    try:
        lat, lon = lugar.split(',')
        return lat, lon
    except ValueError:
        return None, None

# Crear nuevas columnas 'latitud' y 'longitud' usando .loc para evitar el SettingWithCopyWarning
estacionados_cliente[['latitud', 'longitud']] = estacionados_cliente['Lugar_de_inicio'].apply(lambda x: pd.Series(split_coords(x)))
```

```python
# Convertir las coordenadas de 'latitud' y 'longitud' a flotantes
def convert_coord(coord):
    if 'S' in coord or 'W' in coord:
        return -float(coord[:-1])
    elif 'N' in coord or 'E' in coord:
        return float(coord[:-1])
    else:
        return None

estacionados_cliente['latitud'] = estacionados_cliente['latitud'].apply(convert_coord)
estacionados_cliente['longitud'] = estacionados_cliente['longitud'].apply(convert_coord)

# Mostrar el DataFrame después de la conversión
estacionados_cliente.sample(3)
```

```python
# Cambiar la columna latitud por camion_x y longitud por camion_y
estacionados_camion = estacionados_cliente.rename(columns={'latitud': 'camion_x', 'longitud': 'camion_y'})
```

```python
# Filtrar datos con latitud válida
ubi_cliente_clean = ubi_cliente_clean[(ubi_cliente_clean['LATITUD'] >= -90) & (ubi_cliente_clean['LATITUD'] <= 90)]
estacionados_camion_clean = estacionados_camion_clean[(estacionados_camion_clean['camion_x'] >= -90) & (estacionados_camion_clean['camion_x'] <= 90)]

# Obtener las coordenadas en un formato 2D adecuado para sklearn
ubi_cliente_coords = ubi_cliente_clean[['LATITUD', 'LONGITUD']].to_numpy()
camion_coords = estacionados_camion_clean[['camion_x', 'camion_y']].to_numpy()

# Verificar que las coordenadas estén en formato 2D
if ubi_cliente_coords.ndim == 1:
    ubi_cliente_coords = ubi_cliente_coords.reshape(1, -1)
if camion_coords.ndim == 1:
    camion_coords = camion_coords.reshape(1, -1)

# Crear el modelo de vecinos más cercanos
nbrs = NearestNeighbors(n_neighbors=1, algorithm='ball_tree').fit(camion_coords)
distances, indices = nbrs.kneighbors(ubi_cliente_coords)

# Crear el DataFrame vinculado
vinculado_data = []

# Iterar sobre las ubicaciones más cercanas y recopilar la información
for i, index in enumerate(indices):
    ubi_cliente_info = ubi_cliente_clean.iloc[i]
    camion_info = estacionados_camion_clean.iloc[index[0]]
    
    # Coordenadas del cliente y el camión
    cliente_coords = (ubi_cliente_info['LATITUD'], ubi_cliente_info['LONGITUD'])
    camion_coords = (camion_info['camion_x'], camion_info['camion_y'])
    
    # Calcular la distancia haversine en metros
    distancia_metros = geodesic(cliente_coords, camion_coords).meters
    
    vinculado_data.append({
        'CODIGO': ubi_cliente_info['CODIGO'],
        'NOMCLI': ubi_cliente_info['NOMCLI'],
        'LATITUD': ubi_cliente_info['LATITUD'],
        'LONGITUD': ubi_cliente_info['LONGITUD'],
        'Numero_de_placa': camion_info['Numero_de_placa'],
        'Estado_de_viaje': camion_info['Estado_de_viaje'],
        'Tiempo_de_Inicio': camion_info['Tiempo_de_Inicio'],
        'camion_x': camion_info['camion_x'],
        'camion_y': camion_info['camion_y'],
        'Duracion': camion_info['Duracion'],
        'Distancia (grados)': distances[i][0],  # Distancia en grados (como en el código original)
        'Distancia (metros)': distancia_metros  # Nueva columna con la distancia en metros
    })

# Crear un DataFrame con la información vinculada
df_vinculado = pd.DataFrame(vinculado_data)

# Guardar el DataFrame en un archivo Excel
output_file = 'vinculado_datasets_con_distancia_metros.xlsx'
df_vinculado.to_excel(output_file, index=False)

print(f"Archivo Excel guardado como {output_file}")
```
